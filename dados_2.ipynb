c) Qual é a taxa de evasão dos cursos por UF e ano? 

```python
ing=crosstab(index= T['UF'],columns=T['ANO'],values=T['INGRESSANTES'],aggfunc='sum')
con=crosstab(index= T['UF'],columns=T['ANO'],values=T['CONCLUINTES'],aggfunc='sum')
```

```python
C=DataFrame({'uf':T['UF']}).groupby('uf').agg('sum')
x=2020
while x!= 2013:
    C[x]=round(((1-((con[x])/(ing[x-5])))*100),2)
    x-=1
C.drop('Distância',inplace=True)
C.loc['Amapá',2014]=0 # tirando valor negativo
C.loc['Piauí',2018]=0
C
```

```python
arquivos_ies=[]
for arquivo in glob.glob(r'C:\Users\gabri\OneDrive\Documentos\dados_av2_tdadd\ies\*csv'):
     arquivos_ies.append(read_csv(arquivo, sep = ';' , encoding = 'ISO-8859-1',
                              usecols=['NU_ANO_CENSO','SG_IES','CO_IES']))
tabela_ies=concat(arquivos_ies)

tabela_ies.rename(columns = {'NU_ANO_CENSO':'ANO','SG_IES':'SG_IES','CO_IES':'CO_IES'}, inplace = True)
tabela_ies
```

```python
TB_CURSO=tabela[['ANO','UF','CO_IES','INGRESSANTES','CONCLUINTES']]
```

```python
TT=tabela_ies.merge(TB_CURSO,on=['CO_IES','ANO'],how='inner')\
.groupby(['ANO','SG_IES','UF']).agg(
   INGRESSANTES=('INGRESSANTES',sum),CONCLUINTES=('CONCLUINTES',sum))
TT=TT.reset_index().fillna(value={'EVASAO_CURSO':0})
TT
```

```python
anos=[2020,2019,2018,2017,2015,2014]
ar=[]
for i in anos:
    ar.append(TT.query(f'ANO=={i-5}').merge(TT.query(f'ANO=={i}'),on=['SG_IES','UF'],how='inner'))
tab=concat(ar, axis=0)
tab=tab.assign(
    EVASAO=lambda x: round(((1-((x.CONCLUINTES_y)/(x.INGRESSANTES_x)))*100),2)
)
tab=tab[['ANO_y','SG_IES','UF','EVASAO']]
tab.rename(columns={'ANO_y':'ANO'},inplace=True)
tab['EVASAO']=tab['EVASAO'].clip(lower=0,upper=100)
tab
```

d) Qual é a probabilidade de persistência da evasão dos cursos em universidades federais entre 2014 e
2020? Para essa questão, crie uma variável faixa_evasao por quartil dos limiares teóricos para a taxa
evasão

```python

TAB_D=tab_.query('ANO==2014').rename(columns={'EVASAO':'EVASAO_14'}).merge(tab_.query('ANO==2020')\
    .rename(columns={'EVASAO':'EVASAO_20'}),on='SG_IES',how='outer',indicator=True)
fax_eva=TAB_D.assign( quartil_14= lambda z : cut(z.EVASAO_14,bins=[0,25,50,75,100],include_lowest=True),
                quartil_20= lambda z : cut(z.EVASAO_20,bins=[0,25,50,75,100],include_lowest=True)
                )
dd=crosstab(fax_eva.quartil_14,fax_eva.quartil_20,normalize='all')
D=trace(dd)
D
```

e) Qual é a probabilidade de transição para um piora no indicador da evasão dos cursos em universidades
federais entre 2014 e 2020?

```python
E=tril(dd).sum()
E
```
